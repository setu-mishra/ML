```python
#linear regression
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4])  # Feature values
y = np.array([1, 3, 4, 8])  # Target values


X = np.vstack([np.ones(len(x)), x]).T  # X has two columns: 1s (for intercept) and x

theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

# The coefficients are in theta (intercept and slope)
intercept = theta[0]
slope = theta[1]
print(f"Intercept (a0): {intercept}")
print(f"Slope (a1): {slope}")

x_new = np.array([5])  # New value for x
X_new = np.array([[1, x_new[0]]])  # Add the intercept term (1) and reshape the new input as 1x2
y_pred = X_new.dot(theta)  # Calculate the prediction using the model parameters (theta)
print(f"Predicted y for x = 5: {y_pred[0]}")

# Step 4: Visualize the results
plt.scatter(x, y, color='blue', label='Data')  # Scatter plot of the original data points
y_line = X.dot(theta)  # Predictions from the model
plt.plot(x, y_line, color='red', label='Regression Line')  # Plot the regression line

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Different Matrix Method')  # Changed the title here
plt.legend()

# Show the plot
plt.show()
```
